```python
# filter
ar = ['김구','김좌진','안중근','윤봉길',None]

# None값 체크
print(None in ar)

# None 값을 제외한 리스트를 반환
def func1(x):
    return x != None

# filter(함수, 적용할 데이터) // filter를 통해 func1 함수 적용 
ar = list(filter(func1,ar))
print(ar)



def func2(x):
    return len(x)>=3

result = list(filter(func2,ar))
print(result)

# 가 ~ 나

# lambda 사용
ar = ['김구','김좌진','안중근','윤봉길',None]

print(ar)
ar = list(filter(lambda x:x!=None,ar))
print(ar)
result = list(filter(lambda x:len(x)>=3, ar))
print(result)

# ㅇ ~ ㅈ 으로 시작하는 데이터만 뽑기
result = list(filter(lambda x: x[0]>='아' and x[0] <'자',ar))
print(result)
```

```python
# 중첩함수의 이해

def outer():
	outer_data ='외부함수의 데이터'
	
	def inner():
		outer_data ='내부함수의 데이터'
		print(outer_data)
		
		# ver1. outer_data ='내부에서 외부 함수의 데이터 변경'
		# outer_data ='내부에서 외부 함수의 데이터 변경' # ver2. 실행시 에러 // outer_data가 지역변수로써 만들어짐, 
		                                            # 변수를 만드는 문장이 보다 위의 print문에서 없는 outer_data를 실행하기 때문에 에러가 발생한다.
		print(outer_data)

	inner()
	print(outer_data)
	
outer()
	
		
```

```python
# 중첩함수의 이해
# nonlocal
def outer():
	outer_data ='외부함수의 데이터'
	
	def inner():
		# nonlocal 을 명시해줌으로써, 함수 내부에서 데이터를 생성하지 않고, 외부의 데이터를 다시 사용
		nonlocal outer_data 
		print(outer_data)
		
		# ver1. outer_data ='내부에서 외부 함수의 데이터 변경'
		outer_data ='내부에서 외부 함수의 데이터 변경' 
		
		print(outer_data)

	inner()
	print(outer_data)
	
outer()
	
		
```

```python
# 중첩함수의 이해
# global
outer_data ='전역에서 만든 데이터'
def outer():
	outer_data ='외부함수의 데이터'
	
	def inner():
		# nonlocal 을 명시해줌으로써, 함수 내부에서 데이터를 생성하지 않고, 외부의 데이터를 다시 사용
		global outer_data 
		print('1',outer_data)
		
		# ver1. outer_data ='내부에서 외부 함수의 데이터 변경'
		outer_data ='내부에서 외부 함수의 데이터 변경' 
		
		print('2',outer_data)

	inner()
	print('3',outer_data)
	
outer()
	
		
```

```python
import time
def hello():
    print('hello')

hello()


def clock(func):

    def clocked(*args):
        start = time.time()

        result = func(*args)

        end = time.time()
        elapsed = end - start
        print('수행시간:',elapsed)

        # 매개변수 확인
        print('매개변수:,',args)

        # 리턴값
        print('리턴값:',result)

        return result
    
    return clocked




```

```python
from functools import lru_cache

@lru_cache(maxsize=128)
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

# Calling the fibonacci function multiple times
print(fibonacci(5))  # Computes fibonacci(5) and caches the result
print(fibonacci(5))  # Returns the cached result from the previous call
print(fibonacci(10))  # Computes fibonacci(10) and caches the result
print(fibonacci(10))  # Returns the cached result from the previous call
```

```python
class Student:
    # class 내의 instance를 다루기 위함
    def disp(self):
        print("인스턴스 생성")
    
# 인스턴스 생성
# Student() 가 instance 이다.
student = Student()

# 메서드 호출 - bound 호출
# 일반적으로 bound 호출을 사용해라.ㅁ
student.disp()

# 메서드 호출 - unbound 호출
Student.disp(student)
```

```python
class Student:
    # class 내의 instance를 다루기 위함
    def disp(self):
        print("인스턴스 생성")
    
    def setName(self,name):
        self.name = name
        

stu = Student()
stu.setName ='havi'
stu.disp()

stu.score = 94 # instance에 score 라는 속성이 있으면 수정이고, 없으면 생성.
               # score 속성이 없는데도, 다음과 같이 생성할 수 있다.
print(stu.score)
```
