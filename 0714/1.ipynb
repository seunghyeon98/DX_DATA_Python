```python
# IP별 트래픽 합계
# 마지막 항목이 트래픽
from collections  import Counter

ip2traffic=Counter()
with open('./log.txt') as f:
    for line in f:
        ar = line.split()
        try:
            ip = ar[0]
            traffic = int(ar[-1])

            # 카운터 객체에 key값을 부여하고, 값을 누적할땐, += value활용하기
            ip2traffic[ip]+=traffic
        except:
            pass
        
print(dict(ip2traffic))

print(ip2traffic.most_common(10))

```
