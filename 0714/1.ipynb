```python
# IP  
#   
from collections  import Counter

ip2traffic=Counter()
with open('./log.txt') as f:
    for line in f:
        ar = line.split()
        try:
            ip = ar[0]
            traffic = int(ar[-1])

            #   key ,  , += value
            ip2traffic[ip]+=traffic
        except:
            pass
        
print(dict(ip2traffic))

print(ip2traffic.most_common(10))

```
